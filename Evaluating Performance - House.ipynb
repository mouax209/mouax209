```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

import warnings
warnings.filterwarnings('ignore')

postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'houseprices'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))
house_df = pd.read_sql_query('select * from houseprices',con=engine)

# no need for an open connection, as we're only doing a single query
engine.dispose()


house_df.head(10)
```

```python
X = house_df[['overallqual', 'grlivarea', 'garagecars', 'garagearea', 'totalbsmtsf', 'firstflrsf', 'fullbath', 'totrmsabvgrd',  'yearbuilt', 'yearremodadd']]
Y = house_df.saleprice
```

```python
import statsmodels.api as sm
X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

```python
#r-squared: 0.774 and Adj. R-squared:0.772
#yes they are satisfactory because it's not too high that it could be overfitting, but it 
#leaves just enough room for improvement. 
#The AIC and BIC values are pretty much zero, which is good because it's very low
#Because of the values listed above, I think my model is satisfactory - low AIC and BIC with a high R-sqaured
```
