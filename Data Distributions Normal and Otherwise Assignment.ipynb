```python
binomial = np.random.binomial(20, .5, 100)
plt.hist(binomial)
plt.axvline(binomial.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(binomial.mean() + binomial.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(binomial.mean() -binomial.std(), color='r', linestyle='dashed', linewidth=2) 
plt.show
```

```python
normal = np.random.normal(10, 1, 100)
plt.hist(normal)
normal = np.random.normal(5,.5,100)
plt.hist(normal, color='red')
var1 = np.random.normal(10, 1, 100)
var2 = np.random.normal(5, .5, 100)
var3 = var1+var2
mean = np.mean(var3)
sd = np.std(var3)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
```
