```python
from sklearn import neighbors

# Build our model.
knn = neighbors.KNeighborsRegressor(n_neighbors=10)
X = pd.concat([grade_df.G1,grade_df.G2,grade_df.absences], axis=1)
Y = grade_df.G3
knn.fit(X, Y)

```

```python
knn = neighbors.KNeighborsRegressor(n_neighbors=10)
X = pd.concat([grade_df.G1,grade_df.G2,grade_df.absences], axis=1)
Y = grade_df.G3
knn.fit(X, Y)
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(knn, X, Y, cv=10)
print("Unweighted Accuracy: %0.2f (+/- %0.2f)" % (score.mean(), score.std() * 2))
score_w = cross_val_score(knn, X, Y, cv=10)
print("Weighted Accuracy: %0.2f (+/- %0.2f)" % (score_w.mean(), score_w.std() * 2))
```

The models are very similar. The OLS shows r-squared values at 82% and the accuracy rate for the KNN model is at 82% also. I think that in this instance the OLS was a better choice because my variables were based on the theory that together they assume a linear relationship with Y(G3). Therefore, I think that producing the OLS was more telling in helping understand the impact of each feature. I think that the KNN modeling would work best in situations where there aren't assumptions about a direct relationship like if we were forecasting weather as non-meterologists.
