```python
##standarize column names by replacing spaces with underscores and lowercasing all characters
```

```python
df.columns = df.columns.str.replace(' ', '_').str.lower()
df.columns
```

```python
#rename columns 
```

```python
df_cols = ['pmid/pmcid', 'publisher', 'journal_title', 'article_title',
       'cost']
df.columns = df_cols
df.columns
```

```python
##Remove currency sign for easier cal
```

```python
df.cost = df.cost.str.replace('£', '').str.replace('$', '').astype(float) 
df.cost
```

```python
##Answers, lowercase all characters first
```

```python
df['journal_title'] = df.journal_title.str.lower()
most_common_5 = df['journal_title'].value_counts().head(5)
most_common_5
```

```python
journal_title_5=list(most_common_5.index)
journal_title_5
```

```python
df_common_5 = df[df.journal_title.isin(journal_title_5)]
df_common_5
```

```python
df_common_5.groupby('journal_title').agg(['mean', 'median', 'std'])
```
