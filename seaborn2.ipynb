```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
life = sns.load_dataset("life_who")
print(life.head)
```

```python
##continuous variables: pop, percentage expenditure, GDP, BMI
##categorical variable: Status 
```

```python
sns.set(style="white")

g = sns.jointplot(x='Schooling', y='Alcohol',
               data=life, kind="reg", color="g")
plt.suptitle('Country by Years of Schooling and Alcohol Consumption')
plt.subplots_adjust(top=0.9)
plt.show()

g = sns.jointplot(x='Schooling', y='Alcohol',
               data=life, kind="reg", color="g")
plt.suptitle('Country by Years of Schooling and Alcohol Consumption')
plt.subplots_adjust(top=0.9)
plt.show()
```
