```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
from datetime import datetime
start_time = datetime.now()
```

```python
grade_df = pd.read_csv('student-mat.csv')
```

```python
grade_df.head(10)
```

```python
pd.get_dummies(grade_df["Pstatus"])
#A=parents living apart T=parents living together
```

```python
grade_df = pd.concat([grade_df, pd.get_dummies(grade_df['Pstatus'])], axis=1)
```

```python
grade_df
```

```python
grade_df.info()
```

```python
from sklearn import tree
from IPython.display import Image
```

```python
import pydotplus
```

```python
import graphviz
```

```python
from sklearn.model_selection import train_test_split
```

```python
Parents_Apart = grade_df['A']
Parents_Together = grade_df['T']
```

```python
feature_cols= ['absences', 'famrel', 'goout', 'G1', 'G2']
X = grade_df[feature_cols]
```

```python
target_cols = ['A', 'T']
Y = grade_df[target_cols]
```

```python
grade_df[target_cols].shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

```python
#creating decision tree
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf_train = clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#Export to dot format
dot_data = tree.export_graphviz(clf_train, out_file=None, rounded=True, filled=True)
```

```python
print(dot_data)
```

```python
#visualize tree
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
