```python
fill_list = ["PRIMARY_KEY", "STATE", "YEAR", "ENROLL", "TOTAL_REVENUE",
       "FEDERAL_REVENUE", "STATE_REVENUE", "LOCAL_REVENUE",
       "TOTAL_EXPENDITURE", "INSTRUCTION_EXPENDITURE",
       "SUPPORT_SERVICES_EXPENDITURE", "OTHER_EXPENDITURE",
       "CAPITAL_OUTLAY_EXPENDITURE", "GRADES_PK_G", "GRADES_KG_G",
       "GRADES_4_G", "GRADES_8_G", "GRADES_12_G", "GRADES_1_8_G",
       "GRADES_9_12_G", "GRADES_ALL_G", "AVG_MATH_4_SCORE", "AVG_MATH_8_SCORE",
       "AVG_READING_4_SCORE", "AVG_READING_8_SCORE"]
```

```python
education_df.isnull()
```

```python
states = education_df["STATE"].unique()

#interpolate missing values

for state in states:
    education_df.loc[education_df["STATE"] == state, fill_list] = education_df.loc[education_df["STATE"] == state, fill_list].interpolate()
```

```python
#drop missing values after interpolation 
education_df.dropna(inplace=True)
```

```python
#1 Create a new score variable from the weighted averages of all 
#score variables in the datasets. 
```

```python
education_df["overall_score"] = (education_df["GRADES_4_G"]*((education_df["AVG_MATH_4_SCORE"] + education_df["AVG_READING_4_SCORE"])*0.5) + education_df["GRADES_8_G"] * ((education_df["AVG_MATH_8_SCORE"] + education_df["AVG_READING_8_SCORE"])*0.5))/(education_df["GRADES_4_G"] + education_df["GRADES_8_G"])
```

```python
#2 What are the correlations between this newly created score variable and the expenditure 
#types? Which one of the expenditure type is more correlated than the others?
```

```python
education_df[["overall_score","TOTAL_EXPENDITURE", "INSTRUCTION_EXPENDITURE",
       "SUPPORT_SERVICES_EXPENDITURE", "OTHER_EXPENDITURE",
       "CAPITAL_OUTLAY_EXPENDITURE"]].corr()
```

```python
#Overall scores have very little correlation with expenditure. This means that spending does not 
#indicate or forecast higher scores
```

```python
#3 Now, apply PCA to the 4 expenditure types. How much of the total variance is explained 
#by the 1st component?
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
X = education_df[["INSTRUCTION_EXPENDITURE", "SUPPORT_SERVICES_EXPENDITURE",
                  "OTHER_EXPENDITURE", "CAPITAL_OUTLAY_EXPENDITURE"]]

#normalize data
X = StandardScaler().fit_transform(X)
```

```python
sklearn_pca = PCA(n_components=1)
education_df["pca_1"] = sklearn_pca.fit_transform(X)
```

```python
print(
    sklearn_pca.explained_variance_ratio_
)
```

```python
#More than 94% of the total variance is explained by the first principal component
```
