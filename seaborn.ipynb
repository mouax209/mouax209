```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
states = sns.load_dataset("states_all")
print(states.head)
```

```python
##One variable
```

```python
plt.figure(figsize=(70,20))
sns.boxplot(x='INSTRUCTION_EXPENDITURE', data=states)
plt.title('Instruction Expenditure by US DEPT of ED')
plt.show();
```

```python
##PRO: Shows descriptive statistics 
##CON: Many outliers causing it to be ineffective graph for telling a story
```

```python
sns.barplot(x='INSTRUCTION_EXPENDITURE', data=states)
plt.title('Instruction Expenditure by US DEPT of ED')
plt.show();
```

```python
##PRO: Clear 
##CON: Not interesting or shows much information 
```

```python
sns.violinplot(x='INSTRUCTION_EXPENDITURE', data=states)
plt.title('Instruction Expenditure by US DEPT of ED')
plt.show();
```

```python
##PRO: A little better than the barplot 
##CON: Not interesting or shows much information 
```

```python
sns.swarmplot(x='INSTRUCTION_EXPENDITURE', data=states)
plt.title('Instruction Expenditure by US DEPT of ED')
plt.show();
```

```python
##PRO: Much better than the box and violin plots because it shows data points  
##CON: Could use another variable to tell a story 
```

```python
##Two CONTINUOUS Variables
```

```python
g = sns.relplot(x='INSTRUCTION_EXPENDITURE', y='ENROLL', data=states)
plt.title('Instruction Expenditure by Enrollment') 
plt.show()
```

```python
##PRO: Shows a relationship between enrollment and instruction expenditure
##CON: Not interesting, creates some questions about the gaps between data points 
```

```python
g = sns.scatterplot(x='INSTRUCTION_EXPENDITURE', y='ENROLL', data=states)
plt.title('Instruction Expenditure by Enrollment') 
plt.show()
```

```python
##PRO: Holds a little more value with a story to share 
##CON: Not interesting, creates some questions about the gaps between data points 
```

```python
g = sns.lineplot(x='INSTRUCTION_EXPENDITURE', y='ENROLL', data=states)
plt.title('Instruction Expenditure by Enrollment') 
plt.show()
```

```python
##PRO: Like the two previous ones this on 
##CON: Messy and uninteresting 
```

```python
##One Categorical variable and One Continuous variable 
```

```python
plt.figure(figsize=(70,20))
sns.boxplot(x='STATE', y='INSTRUCTION_EXPENDITURE', data=states)
plt.title('Instruction Expenditure by State') 
plt.show()
```

```python
##PRO: A lot of information about each state's spending
##CON: Not interesting, creates some questions about the gaps between data points 
```

```python
plt.figure(figsize=(40,10))
sns.stripplot(x='STATE', y='INSTRUCTION_EXPENDITURE', data=states)
plt.title('Instruction Expenditure by State') 
plt.show()
```

```python
##PRO: A lot of information about each state's spending
##CON: Visual is not very clear or easy to understand
```

```python
plt.figure(figsize=(40,10))
sns.boxenplot(x='STATE', y='INSTRUCTION_EXPENDITURE', data=states)
plt.title('Instruction Expenditure by State') 
plt.show()
```

```python
##PRO: A lot of information about each state's spending
##CON: Still a lot of information to sort through 
```

```python
plt.figure(figsize=(40,10))
sns.swarmplot(x='STATE', y='INSTRUCTION_EXPENDITURE', data=states)
plt.title('Instruction Expenditure by State') 
plt.show()
```

```python
##PRO: A lot of information 
##CON: less visually appealing and informative as the box/boxen plots 
```

```python
plt.figure(figsize=(40,10))
sns.barplot(x='STATE', y='INSTRUCTION_EXPENDITURE', data=states)
plt.title('Instruction Expenditure by State') 
plt.show()
```

```python
##PRO: Very clear and easy on the eyes
##CON: Needs to be tailored better
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x='STATE', y='INSTRUCTION_EXPENDITURE', data=states)
plt.title('Instruction Expenditure by State') 
plt.show()
```

```python
##PRO: Data is aggregated and not sparse with data points
##CON: Needs explanation on how to interpret violin graphs
```
